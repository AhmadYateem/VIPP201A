### **Intro to TensorFlow for Deep Learning - Celsius to Fahrenheit**

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
celsius = np.array([13,47,2,-1,8,0,9,12,32] , dtype= float)
fahrenheit = np.array([55.4, 116.6, 35.6, 30.2 , 46.4, 32, 48.2 , 53.6, 89.6] , dtype = float)
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])
```

```python
model.compile(loss="mean_squared_error" , optimizer = tf.keras.optimizers.Adam(0.01))
```

```python
training = model.fit(celsius, fahrenheit, epochs = 6000, verbose = False)
print("Training over")
```

```python
plt.xlabel("Epoch Number")
plt.ylabel("Loss")
plt.plot(training.history["loss"])
```

```python
celsius_v = np.array([20 , 90 , -3 , 4 , 28 , 7 , 56 , 22 , 1 ] , dtype = float)
fahrenheit_v = np.array([68 , 194 , 26.6, 39.2, 82.4 , 44.6 , 132.8, 71.6 , 33.8  ] , dtype = float)
predictions = model.predict(celsius_v).flatten()
for i in range(len(predictions)):
  print("Prediction:" , predictions[i] , "\nActual value:", fahrenheit_v[i], end="\n\n")
loss = model.evaluate(celsius_v, fahrenheit_v)
print("Loss:",loss)
```

```python
plt.figure(figsize=(8, 5))

plt.plot(celsius_v, fahrenheit_v, color='orange', label='Actual Values', marker='o')
plt.plot(celsius_v, predictions, color='blue', label='Predicted Values', marker='x')

plt.title("Predictions vs Actual Values")
plt.xlabel("Celsius")
plt.ylabel("Fahrenheit")

plt.legend()
plt.show()
```
